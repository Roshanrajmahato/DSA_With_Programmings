# Question 1 → Maximum Sum Subarray of Size K



```python
def max_sum_subarray(arr, k):

    window_sum = 0
    max_sum = 0

    # First window
    for i in range(k):
        window_sum += arr[i]

    max_sum = window_sum

    # Slide window
    for i in range(k, len(arr)):

        window_sum += arr[i]       # Add new element
        window_sum -= arr[i-k]     # Remove old element

        if window_sum > max_sum:
            max_sum = window_sum

    return max_sum


arr = [2, 1, 5, 1, 3, 2]
k = 3

print(max_sum_subarray(arr, k))
```

### Insight

1. **First Create One Window**
2. **Then Move Window One By One**
3. **One Element Add**
4. **One Element Remove**
5. **Current Window Sum Continous Updating**
6. **Final Maximum Updating**

# Question 2 → Smallest Subarray With Given Sum

```python
def smallest_subarray(arr, target):

    left = 0
    current_sum = 0
    min_length = float('inf')

    for right in range(len(arr)):

        current_sum += arr[right]

        while current_sum >= target:

            min_length = min(min_length, right-left+1)

            current_sum -= arr[left]
            left += 1

    return min_length


arr = [2, 1, 5, 2, 3, 2]
target = 7

print(smallest_subarray(arr, target))
```

### Insight

1. **Expand Window Until Condition Satisfy**
2. **When Condition Satisfy, Try To Shrink Window**
3. **Find Minimum Length**

def longest_unique(s):

    left = 0
    seen = set()

    max_len = 0

    for right in range(len(s)):

#  **Sliding Window** problem (FIXED window size).

## Your code is finding the **maximum sum of a subarray of size `k`**.
## 🧠 **Question**

**Given an array of integers `nums` and an integer `k`, find the maximum sum of any contiguous subarray of size `k`.**

### 📌 Example:

```
Input:
nums = [2, 1, 5, 1, 3, 2]
k = 3

Output:
9

Explanation:
Subarrays of size 3:
[2,1,5] → sum = 8  
[1,5,1] → sum = 7  
[5,1,3] → sum = 9  ✅ (maximum)  
[1,3,2] → sum = 6  
```

````python
def max_sum_subarray(nums, k):
    i = 0
    j = 0
    n = len(nums)
    current_sum = 0
    max_sum = float('-inf')

    while j < n:
        current_sum += nums[j]

        # Window size < k # expending the window till size k (condition)
        if (j - i + 1) < k:
            j += 1

        # Window size == k # Mainting te window size to k
        elif (j - i + 1) == k:
            max_sum = max(max_sum, current_sum)

            # Slide the window
            current_sum -= nums[i]
            i += 1
            j += 1

    return max_sum
````

#  **Sliding Window** problem (variable window size).

**Given an array of positive integers `arr` and an integer `k`, find the length of the longest contiguous subarray whose sum is equal to `k`.**

---

## 📌 Example

```id="ex1"
Input:
arr = [4, 1, 1, 1, 2, 3, 5]
k = 5

Output:
4

Explanation:
Subarray [1,1,1,2] → sum = 5 → length = 4 ✅ (maximum)
```

---


````python

## 🧩 **Python Solution (Sliding Window)**
def longest_subarray_with_sum_k(arr, k):
    n = len(arr)
    i = 0
    j = 0
    max_len = 0
    current_sum = 0

    while j < n:
        current_sum += arr[j]

        # Case 1: sum < k → expand window
        if current_sum < k:
            j += 1

        # Case 2: sum == k → update answer
        elif current_sum == k:
            max_len = max(max_len, j - i + 1)
            j += 1

        # Case 3: sum > k → shrink window
        else:
            while current_sum > k and i <= j:
                current_sum -= arr[i]
                i += 1

            if current_sum == k:
                max_len = max(max_len, j - i + 1)

            j += 1

    return max_len

````

````python
def longest_subarray_with_sum_k_neg(arr, k):
    prefix_sum_map = {}
    current_sum = 0
    max_len = 0

    for i in range(len(arr)):
        current_sum += arr[i]

        # Case 1: sum itself equals k
        if current_sum == k:
            max_len = i + 1

        # Case 2: check if (current_sum - k) exists
        if (current_sum - k) in prefix_sum_map:
            max_len = max(max_len, i - prefix_sum_map[current_sum - k])

        # Store first occurrence of prefix sum
        if current_sum not in prefix_sum_map:
            prefix_sum_map[current_sum] = i

    return max_len

````
