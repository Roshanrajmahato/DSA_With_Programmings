# Question 2 → Two Sum

```python id="0n0c2v"
def two_sum(arr, target):

    hashmap = {}

    for i in range(len(arr)):

        needed = target - arr[i]

        if needed in hashmap:
            return (hashmap[needed], i)

        hashmap[arr[i]] = i

    return "No Pair"


arr = [2,7,11,15]
target = 9

print(two_sum(arr, target))
```

### Insight

1. **Current Element Need Another Value**
2. **Find Needed Value Using Target**
3. **Check Previous Stored Elements**
4. **If Found → Pair Ready**
5. **Else Store Current Element**

    for num in arr:

        if num in seen:

# Question 5 → Longest Consecutive Sequence

```python id="zz8h87"
def longest_consecutive(arr):

    nums = set(arr)

    longest = 0

    for num in nums:

        if num - 1 not in nums:

            current = num
            streak = 1

            while current + 1 in nums:

                current += 1
                streak += 1

            longest = max(longest, streak)

    return longest


arr = [100,4,200,1,3,2]

print(longest_consecutive(arr))
```

### Insight

1. **Convert Array Into Set For Fast Lookup**
2. **Start Sequence Only From Starting Point**
3. **Keep Checking Next Consecutive Number**
4. **Current Streak Continous Updating**
5. **Maximum Sequence Continous Updating**

# Find first non Repeating character

````python
def first_non_repeating_char(s):
    dic = {}

    # Count frequency
    for ch in s:
        if ch not in dic:
            dic[ch] = 1
        else:
            dic[ch] += 1

    # Using dic.items()
    for key, val in dic.items():
        if val == 1:
            return key

    return -1


# Example
s = "RoshanRajMahato"
print(first_non_repeating_char(s))
````

# Most Frequent Element — Optimized (Dictionary)

````python
def mostFrequentElement(arr):

    freq = {}

    for num in arr:
        if num in freq:
            freq[num] += 1
        else:
            freq[num] = 1


    maxCount = 0
    mostFrequent = -1

    for key in freq:
        if freq[key] > maxCount:
            maxCount = freq[key]
            mostFrequent = key

    return mostFrequent


arr = [1, 2, 3, 2, 5, 2, 3]

print("Most Frequent Element:", mostFrequentElement(arr))
````

# Find First non-repeating character index || All non-repeating characters


````python
def non_repeating_info(s):
    dic = {}

    # Case-insensitive counting
    for ch in s:
        ch_lower = ch.lower()
        if ch_lower not in dic:
            dic[ch_lower] = 1
        else:
            dic[ch_lower] += 1

    # First non-repeating character index
    first_index = -1
    for i, ch in enumerate(s):
        if dic[ch.lower()] == 1:
            first_index = i
            break

    # All non-repeating characters (original form)
    all_chars = []
    for ch in s:
        if dic[ch.lower()] == 1:
            all_chars.append(ch)

    return first_index, all_chars


# Example
s = "RoshanRajMahato"
index, chars = non_repeating_info(s)

print("First Non-Repeating Index:", index)
print("All Non-Repeating Characters:", chars)
````

# Rank of the Elements in the Array Replace on sorting


````python
def replaceWithRank(arr):
    rank_map = {}
    rank = 1
    for num in sorted(arr):
        if num not in rank_map:
            rank_map[num] = rank
            rank += 1
    return [rank_map[num] for num in arr]
try:
    # n = int(input().strip())
    # arr = list(map(int, input().split()))
    arr=[30,20,10,40,50]
    result = replaceWithRank(arr)
    print(*result)
except EOFError:
    pass
````

# Group Anagrams


````python
words = ["eat","tea","tan","ate","nat","bat"]
groups = {}
for word in words:
    key = ''.join(sorted(word))
    if key not in groups:
        groups[key] = []
    groups[key].append(word)
print(list(groups.values()))
````

# Remove Characters from First String Present in Second String


````python
def removeChars(str1, str2):
    # Create a set for fast lookup
    remove_set = set(str2)
    result = []
    
    for ch in str1:
        if ch not in remove_set:
            result.append(ch)
            
    return "".join(result)

str1 = input().strip()
str2 = input().strip()
print(removeChars(str1, str2))
````

# Find All Repeating Numbers

````python
def allRepeatingNumbers(arr):

    seen = set()
    repeating = set()

    for num in arr:

        if num in seen:
            repeating.add(num)

        else:
            seen.add(num)

    return list(repeating)


arr = [1, 2, 3, 2, 5, 3]

print("Repeating Numbers:", allRepeatingNumbers(arr))
````
