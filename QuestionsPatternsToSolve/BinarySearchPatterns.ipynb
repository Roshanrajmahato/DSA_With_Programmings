# Question 1 → Simple Binary Search

```python id="n1jrbp"
def binary_search(arr, target):

    left = 0
    right = len(arr) - 1

    while left <= right:

        mid = (left + right) // 2

        if arr[mid] == target:
            return mid

        elif arr[mid] < target:
            left = mid + 1

        else:
            right = mid - 1

    return -1


arr = [1,2,3,4,5,6,7]
target = 5

print(binary_search(arr, target))
```

### Insight

1. **Find Mid Element**
2. **Compare With Target**
3. **If Target Bigger → Ignore Left Half**
4. **If Target Smaller → Ignore Right Half**
5. **Search Space Continous Reducing**

---

# Question 2 → First Occurrence Of Element

```python id="v0w2x5"
def first_occurrence(arr, target):

    left = 0
    right = len(arr) - 1

    ans = -1

    while left <= right:

        mid = (left + right) // 2

        if arr[mid] == target:

            ans = mid
            right = mid - 1

        elif arr[mid] < target:
            left = mid + 1

        else:
            right = mid - 1

    return ans


arr = [1,2,2,2,3,4]
target = 2

print(first_occurrence(arr, target))
```

### Insight

1. **When Match Found, Do Not Stop**
2. **Store Current Answer**
3. **Move Toward Left Side**
4. **Try To Find Earlier Occurrence**

### Insight

1. **Search Space From 0 To N**
2. **Check Mid Square**
3. **If Mid Square Smaller → Move Right**
4. **Else Move Left**
5. **Best Possible Answer Continous Updating**
