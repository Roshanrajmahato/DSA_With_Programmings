# Question 1 → Reverse String

```python id="4x8mza"
def reverse_string(s):

    ans = ""

    for i in range(len(s)-1, -1, -1):

        ans += s[i]

    return ans


s = "ROSHAN"

print(reverse_string(s))
```

### Insight

1. **Traverse String From Backward**
2. **Add Characters One By One**
3. **New String Build In Reverse Order**
4. **Character Access Using Index**

---

# Question 2 → Palindrome Check

```python id="3p1kqv"
def palindrome(s):

    left = 0

    right = len(s) - 1

    while left < right:

        if s[left] != s[right]:
            return False

        left += 1
        right -= 1

    return True


s = "madam"

print(palindrome(s))
```

### Insight

1. **Compare Both Side Characters**
2. **If Characters Mismatch → Not Palindrome**
3. **Move Toward Center**
4. **Symmetry Checking Logic**

# Question 3 → Frequency Of Characters

```python id="0v2ywx"
def frequency(s):

    freq = {}

    for ch in s:

        if ch in freq:

            freq[ch] += 1

        else:

            freq[ch] = 1

    return freq


s = "programming"

print(frequency(s))
```

### Insight

1. **Each Character Become Key**
2. **Count Frequency**
3. **If Already Present → Increase Count**
4. **Dictionary Helps Fast Storage**

---

# Question 4 → Longest Common Prefix

```python id="6n8wpt"
def longest_prefix(arr):

    prefix = arr[0]

    for word in arr[1:]:

        while word[:len(prefix)] != prefix:

            prefix = prefix[:-1]

    return prefix


arr = ["flower","flow","flight"]

print(longest_prefix(arr))
```

### Insight

1. **Start With First Word As Prefix**
2. **Compare With Other Words**
3. **If Mismatch → Reduce Prefix**
4. **Continue Until Common Prefix Found**

# Question 5 → Anagram Check

```python id="5k0rzn"
def anagram(s1, s2):

    if len(s1) != len(s2):
        return False

    freq = {}

    for ch in s1:

        freq[ch] = freq.get(ch, 0) + 1

    for ch in s2:

        if ch not in freq:
            return False

        freq[ch] -= 1

        if freq[ch] < 0:
            return False

    return True


s1 = "listen"
s2 = "silent"

print(anagram(s1, s2))
```

### Insight

1. **Both Strings Must Have Same Characters**
2. **Count Frequency Of First String**
3. **Decrease Frequency Using Second String**
4. **If Any Character Missing/Mismatch → Not Anagram**
5. **Frequency Matching Main Logic**
