# Python Problem: Find Sum of Elements in Array

```python
def sumOfArray(arr):
    total = 0
    for num in arr:
        total = total + num
    return total


arr = [1, 4, 5, 6, 1]
print("Sum:", sumOfArray(arr))
```

# Python Problem: Find Maximum Element in Array

```python
def maxOfArray(arr):
    ans = arr[0]   # Start with first element
    for i in range(len(arr)):
        if arr[i] > ans:
            ans = arr[i]
    return ans


arr = [10, 5, 3, 6, 2, 8, 4]
result = maxOfArray(arr)
print("Max:", result)
```

# PROBLEM PYTHON 6: Rotate List to the Left by N Times

```python
l1 = [11, 22, 33, 44, 55]

n = int(input("Enter a number: "))

for _ in range(n):

    temp = l1[0]

    for i in range(len(l1) - 1):
        l1[i] = l1[i + 1]

    l1[len(l1) - 1] = temp

    print(l1)
```

```python
l1 = [11, 22, 33, 44, 55]
n = int(input("Enter a number: "))
for _ in range(n):
    temp = l1[len(l1) - 1]
    # Shift elements to right
    for i in range(len(l1) - 1, 0, -1):
        l1[i] = l1[i - 1]
    l1[0] = temp
    print(l1)
```

# Python Problem: Check if Array is Sorted

```python
def isSorted(arr):
    check = True
    for i in range(1, len(arr)):
        if arr[i] < arr[i - 1]:
            check = False
            break
    return check


arr = [1, 2, 3, 4, 5]
result = isSorted(arr)
print("Is Sorted:", result)
```

# The Problem: Product of List Excluding Self

```python
def product_except_self_division(l1):
    total_product = 1
    for x in l1:
        total_product *= x
    
    # Use floor division (//) to keep results as integers
    res = [total_product // i for i in l1]
    return res

l1 = [2, 4, 5, 6, 8]
print(f"Output using division: {product_except_self_division(l1)}")
```

# The Problem: Find Leaders in an Array

```python
def find_leaders(arr):
    size = len(arr)
    if size == 0:
        return []

    # The rightmost element is always a leader
    leaders = []
    maxright = arr[size - 1]
    leaders.append(maxright)

    # Scan the array from right to left (starting from second to last)
    for i in range(size - 2, -1, -1):
        if arr[i] > maxright:
            leaders.append(arr[i])
            maxright = arr[i]

    # The leaders are collected in reverse order, reverse them from back
    return leaders[::-1]

l1 = [16, 17, 4, 3, 5, 2]
print(f"The leaders are: {find_leaders(l1)}")
```

### Inlace Methods

```python
def move_zeros_to_end(arr):
    count = 0  # Position for next non-zero

    # Move non-zero elements forward
    for i in range(len(arr)):
        if arr[i] != 0:
            arr[count] = arr[i]
            count += 1

    # Fill remaining positions with zeros
    while count < len(arr):
        arr[count] = 0
        count += 1

    return arr


# Example
arr = [4, 5, 0, 1, 9, 0, 5, 0]
print(move_zeros_to_end(arr))
```

# The Problem: First and Last Index of Target

```python
def find_first_and_last(nums, target):
    first_idx = -1
    last_idx = -1

    # Single pass through the list
    for i in range(len(nums)):
        if nums[i] == target:
            # If we haven't found it yet, this is the first occurrence
            if first_idx == -1:
                first_idx = i
            
            # Always update the last_idx; the final update will be the true last
            last_idx = i

    return (first_idx, last_idx)

# Data from your notes
l1 = [2, 3, 1, 4, 2, 5, 6, 7, 1, 2, 4, 8, 1, 4, 5]
target_val = int(input("Enter the target: "))

result = find_first_and_last(l1, target_val)
print(f"Starting and ending index: {result}")
```

# Python Problem: Find First and Last Occurrence of Element in Array

```python
def firstOccurrence(arr, x):

    for i in range(len(arr)):
        if arr[i] == x:
            return i

    return -1


def lastOccurrence(arr, x):

    for i in range(len(arr) - 1, -1, -1):
        if arr[i] == x:
            return i

    return -1


arr = [1, 2, 3, 2, 5, 4, 2, 8, 9]
x = 2

print("First Occurrence Index:", firstOccurrence(arr, x))
print("Last Occurrence Index:", lastOccurrence(arr, x))
```
