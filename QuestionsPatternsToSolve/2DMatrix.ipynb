# Question 1 → Matrix Traversal


```python

def traversal(matrix):

    rows = len(matrix)

    cols = len(matrix[0])

    for i in range(rows):

        for j in range(cols):

            print(matrix[i][j], end=" ")

        print()


matrix = [
    [1,2,3],
    [4,5,6],
    [7,8,9]
]

traversal(matrix)
```

# Question 2 → Row Sum Of Matrix


```python
def row_sum(matrix):

    for row in matrix:

        current_sum = 0

        for num in row:

            current_sum += num

        print(current_sum)


matrix = [
    [1,2,3],
    [4,5,6],
    [7,8,9]
]

row_sum(matrix)
```

# Question 3 → Transpose Of Matrix


```python
def transpose(matrix):

    rows = len(matrix)

    cols = len(matrix[0])

    result = [[0]*rows for _ in range(cols)]

    for i in range(rows):

        for j in range(cols):

            result[j][i] = matrix[i][j]

    return result


matrix = [
    [1,2,3],
    [4,5,6]
]

print(transpose(matrix))
```

```python
# Inplace for only square Matrix
def inplacetranspose(matrix):
    row=len(matrix)
    col=len(matrix[0])
    for r in range(row):
        for c in range(col):
            if r<c:
                matrix[r][c],matrix[c][r]=matrix[c][r],matrix[r][c]

    return matrix

 
def transpose(matrix):
    row=len(matrix)
    col=len(matrix[0])
    newmatrix=[[0]*row for _ in range(col)]
    for i in range(col):
        for j in range(row):
            newmatrix[i][j]=matrix[j][i]
    
    return newmatrix

matrix=[
 [1,4,7] ,
 [2,5,8],
 [3,6,9]
]
print(f"Normal Transpose {transpose(matrix)}")

print()

print(f"Inplace Tranpose Is {inplacetranspose(matrix)}")

```

# Question 4 → Spiral Traversal


```python
def spiral(matrix):

    top = 0
    bottom = len(matrix) - 1

    left = 0
    right = len(matrix[0]) - 1

    result = []

    while top <= bottom and left <= right:

        for i in range(left, right + 1):

            result.append(matrix[top][i])

        top += 1

        for i in range(top, bottom + 1):

            result.append(matrix[i][right])

        right -= 1

        if top <= bottom:

            for i in range(right, left - 1, -1):

                result.append(matrix[bottom][i])

            bottom -= 1

        if left <= right:

            for i in range(bottom, top - 1, -1):

                result.append(matrix[i][left])

            left += 1

    return result


matrix = [
    [1,2,3],
    [4,5,6],
    [7,8,9]
]

print(spiral(matrix))
```


---
### Insight

1. **Traverse Layer By Layer**
2. **Move Right → Down → Left → Up**
3. **After One Boundary Complete**
4. **Shrink Boundaries**
5. **Continue Until Matrix Ends**

# Question 5 → Search In Sorted Matrix


```python
def search_matrix(matrix, target):

    rows = len(matrix)

    cols = len(matrix[0])

    row = 0
    col = cols - 1

    while row < rows and col >= 0:

        current = matrix[row][col]

        if current == target:
            return True

        elif current > target:
            col -= 1

        else:
            row += 1

    return False


matrix = [
    [1,4,7],
    [2,5,8],
    [3,6,9]
]

print(search_matrix(matrix, 5))
```


---
### Insight

1. **Start From Top Right**
2. **If Current Bigger → Move Left**
3. **If Current Smaller → Move Down**
4. **Search Space Continous Reducing**
5. **Matrix Sorted Property Used Efficiently**
