# The Problem: Find the Majority Element

```python
def find_majority(nums):
    n = len(nums)
    
    # Outer loop picks each number one by one
    for i in range(n):
        count = 0
        
        # Inner loop counts how many times nums[i] appears in the whole list
        for j in range(n):
            if nums[j] == nums[i]:
                count += 1
        
        # Check if this count is the majority
        if count > n / 2:
            return nums[i]
            
    return -1

# Test
l1 = [1, 2, 1, 3, 1, 1, 4, 1]
print(f"Majority Element: {find_majority(l1)}")
```

# The Problem: Maximum Stock Profit

```python
def stock_exchange(prices):
    # Check if there are enough days to trade
    if len(prices) < 2:
        return "There is no proper stock"

    max_profit = 0
    
    # Loop through each day to buy
    for buy in range(len(prices)):
        # Loop through each future day to sell
        for sell in range(buy + 1, len(prices)):
            current_profit = prices[sell] - prices[buy]
            
            # Update max_profit if current deal is better
            if max_profit < current_profit:
                max_profit = current_profit
                
    return max_profit

prices=[10, 7, 1, 3, 5, 6, 9]
maxPofit=stock_exchange(prices)
print(maxPofit)
```

# Python Problem: Find Unique Element in Array

```python
def findUnique(arr):
    n = len(arr)

    for i in range(n):
        for j in range(i + 1, n):
            if arr[i] == arr[j]:
                arr[i] = -1
                arr[j] = -1
    ans = -1

    for i in range(n):
        if arr[i] > 0: 
            ans = arr[i]
            
    return ans

n = int(input("Enter array size: "))
arr = []
print("Enter", n, "elements")

for i in range(n):
    arr.append(int(input()))
print("Unique element:", findUnique(arr))
```

# Python Problem: Count Pair Sum Equal to Target

```python
def pairSum(arr, target):
    n = len(arr)
    ans = 0

    for i in range(n):  # first number
        for j in range(i + 1, n):  # second number
            if arr[i] + arr[j] == target:
                ans += 1

    return ans


# Main execution
n = int(input("Enter array size: "))

arr = []

print("Enter", n, "elements")

for i in range(n):
    arr.append(int(input()))

target = int(input("Enter target sum: "))

print("Pair Count:", pairSum(arr, target))
```

# Python Problem: Find First Repeating Number in Array

```python
def firstRepeatingNumber(arr):
    n = len(arr)

    for i in range(n):  # first number
        for j in range(i + 1, n):  # second number
            if arr[i] == arr[j]:  # found answer
                return arr[i]

    return -1


# Main execution
n = int(input("Enter array size: "))

arr = []

print("Enter", n, "elements")

for i in range(n):
    arr.append(int(input()))

print("First Repeating Number:", firstRepeatingNumber(arr))
```

# Python Problem: Count Triplets with Sum Equal to Target

```python
def tripletSum(arr, target):

    ans = 0
    n = len(arr)

    for i in range(n):  # first number

        for j in range(i + 1, n):  # second number

            for k in range(j + 1, n):  # third number

                if arr[i] + arr[j] + arr[k] == target:

                    ans += 1

    return ans


# Main execution
n = int(input("Enter array size: "))

arr = []

print("Enter", n, "elements")

for i in range(n):
    arr.append(int(input()))

target = int(input("Enter target sum: "))

print("Triplet Count:", tripletSum(arr, target))
```
