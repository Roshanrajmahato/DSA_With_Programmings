# Question 1 → Build Prefix Sum Array

```python id="0u1y7m"
def prefix_sum(arr):

    prefix = [0] * len(arr)

    prefix[0] = arr[0]

    for i in range(1, len(arr)):

        prefix[i] = prefix[i-1] + arr[i]

    return prefix


arr = [2,4,6,8]

print(prefix_sum(arr))
```

### Insight

1. **Current Prefix Depends On Previous Prefix**
2. **Store Running Sum**
3. **Avoid Recalculating Total Again**
4. **Each Position Contains Sum Till That Index**

---

# Question 2 → Range Sum Query

```python id="4x75wf"
def range_sum(arr, left, right):

    prefix = [0] * len(arr)

    prefix[0] = arr[0]

    for i in range(1, len(arr)):

        prefix[i] = prefix[i-1] + arr[i]

    if left == 0:
        return prefix[right]

    return prefix[right] - prefix[left-1]


arr = [2,4,6,8,10]

print(range_sum(arr, 1, 3))
```

### Insight

1. **Store Prefix Sum First**
2. **Range Sum = Bigger Prefix - Smaller Prefix**
3. **No Need To Traverse Again**
4. **Very Fast Range Calculation**

# Question 3 → Running Sum

```python id="2ix2ra"
def running_sum(arr):

    result = []

    current_sum = 0

    for num in arr:

        current_sum += num

        result.append(current_sum)

    return result


arr = [1,2,3,4]

print(running_sum(arr))
```

### Insight

1. **One Variable Maintain Running Total**
2. **Each Step Add Current Element**
3. **Store Updated Result**
4. **Current Result Depends On Previous Result**

---

# Question 4 → Count Subarrays With Given Sum

```python id="p6skmo"
def count_subarrays(arr, target):

    count = 0

    for i in range(len(arr)):

        current_sum = 0

        for j in range(i, len(arr)):

            current_sum += arr[j]

            if current_sum == target:
                count += 1

    return count


arr = [1,2,3,2,1]
target = 5

print(count_subarrays(arr, target))
```

### Insight

1. **Start One Subarray**
2. **Keep Adding Elements**
3. **Running Sum Continous Updating**
4. **If Sum Match → Count Increase**

# Question 5 → Equilibrium Index

```python id="ax0x3m"
def equilibrium(arr):

    total_sum = sum(arr)

    left_sum = 0

    for i in range(len(arr)):

        right_sum = total_sum - left_sum - arr[i]

        if left_sum == right_sum:
            return i

        left_sum += arr[i]

    return -1


arr = [1,3,5,2,2]

print(equilibrium(arr))
```

### Insight

1. **Total Sum Already Known**
2. **Left Sum Continous Updating**
3. **Right Sum Calculate Using Formula**
4. **Check Balance Point**
5. **If Left Sum == Right Sum → Equilibrium Found**
