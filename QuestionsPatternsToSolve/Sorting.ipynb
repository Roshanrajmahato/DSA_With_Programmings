# Question 1 → Bubble Sort

```python id="2v8k0q"
def bubble_sort(arr):

    n = len(arr)

    for i in range(n):

        for j in range(0, n-i-1):

            if arr[j] > arr[j+1]:

                arr[j], arr[j+1] = arr[j+1], arr[j]

    return arr


arr = [5,1,4,2,8]

print(bubble_sort(arr))
```

### Insight

1. **Compare Adjacent Elements**
2. **If Wrong Order → Swap**
3. **Largest Element Bubble To End**
4. **After Every Pass One Position Fixed**
5. **Repeated Comparison & Swapping**

---

# Question 2 → Selection Sort

```python id="7m2xqa"
def selection_sort(arr):

    n = len(arr)

    for i in range(n):

        min_index = i

        for j in range(i+1, n):

            if arr[j] < arr[min_index]:

                min_index = j

        arr[i], arr[min_index] = arr[min_index], arr[i]

    return arr


arr = [64,25,12,22,11]

print(selection_sort(arr))
```

### Insight

1. **Find Minimum Element**
2. **Place Minimum At Correct Position**
3. **One Position Fixed After Each Pass**
4. **Selection Of Best Candidate Happens**

# Question 3 → Insertion Sort

```python id="6r0ywp"
def insertion_sort(arr):

    for i in range(1, len(arr)):

        current = arr[i]

        j = i - 1

        while j >= 0 and arr[j] > current:

            arr[j+1] = arr[j]

            j -= 1

        arr[j+1] = current

    return arr


arr = [5,2,4,6,1,3]

print(insertion_sort(arr))
```

### Insight

1. **Left Side Already Sorted**
2. **Take Current Element**
3. **Shift Bigger Elements**
4. **Insert Current Element At Correct Position**
5. **Gradually Sorted Portion Increases**

---

# Question 4 → Merge Sort

```python id="9k1vzt"
def merge_sort(arr):

    if len(arr) <= 1:
        return arr

    mid = len(arr) // 2

    left = merge_sort(arr[:mid])

    right = merge_sort(arr[mid:])

    return merge(left, right)


def merge(left, right):

    result = []

    i = 0
    j = 0

    while i < len(left) and j < len(right):

        if left[i] < right[j]:

            result.append(left[i])

            i += 1

        else:

            result.append(right[j])

            j += 1

    result.extend(left[i:])

    result.extend(right[j:])

    return result


arr = [38,27,43,3,9,82,10]

print(merge_sort(arr))
```

### Insight

1. **Divide Array Into Smaller Parts**
2. **Sort Smaller Parts**
3. **Merge Sorted Parts Together**
4. **Divide And Conquer Logic**
5. **Recursive Splitting Happens**

# Question 5 → Quick Sort

```python id="8v5m2p"
def quick_sort(arr):

    if len(arr) <= 1:
        return arr

    pivot = arr[-1]

    left = []
    right = []

    for num in arr[:-1]:

        if num < pivot:

            left.append(num)

        else:

            right.append(num)

    return quick_sort(left) + [pivot] + quick_sort(right)


arr = [10,7,8,9,1,5]

print(quick_sort(arr))
```

### Insight

1. **Choose Pivot Element**
2. **Smaller Elements Go Left**
3. **Bigger Elements Go Right**
4. **Recursively Sort Both Sides**
5. **Pivot Reaches Correct Position Automatically**
