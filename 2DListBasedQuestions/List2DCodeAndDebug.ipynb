# 📘 Set Matrix Zeroes — LeetCode 73


🔗 **Problem Link:**
[https://leetcode.com/problems/set-matrix-zeroes/](https://leetcode.com/problems/set-matrix-zeroes/)

📚 **Source Reference:**
[https://codeanddebug.in/blog/set-matrix-zeroes-leetcode-73/](https://codeanddebug.in/blog/set-matrix-zeroes-leetcode-73/)

---

### 🧩 Problem Statement

Given an **m × n matrix**, if an element is **0**, set its **entire row and column** to **0**.

You must do it **in-place**.

---

### 🧪 Example 1

```text
Input:
matrix =
[
 [1,1,1],
 [1,0,1],
 [1,1,1]
]

Output:
[
 [1,0,1],
 [0,0,0],
 [1,0,1]
]
```

---

### 🧪 Example 2

```text
Input:
matrix =
[
 [0,1,2,0],
 [3,4,5,2],
 [1,3,1,5]
]

Output:
[
 [0,0,0,0],
 [0,4,5,0],
 [0,3,1,0]
]
```

---

### 🔒 Constraints

```text
m == matrix.length
n == matrix[0].length

1 <= m, n <= 200
-2^31 <= matrix[i][j] <= 2^31 - 1
```

---



### 🧠 METHOD 1: Brute Force (Using Special Marker)



---

### 📌 Approach

Steps:

1. Traverse matrix
2. If **0 found**, mark row & column using **infinity**
3. Do not overwrite original zeros
4. Convert infinity → 0

Key Idea:

```text
Use float("inf") as temporary marker
```

---



### 💻 Code


````python
from typing import List

class Solution:

    def markInfinity(self, matrix, row, col):

        r = len(matrix)
        c = len(matrix[0])

        # Mark column
        for i in range(r):
            if matrix[i][col] != 0:
                matrix[i][col] = float("inf")

        # Mark row
        for j in range(c):
            if matrix[row][j] != 0:
                matrix[row][j] = float("inf")

    def setZeroes(self, matrix: List[List[int]]) -> None:

        r = len(matrix)
        c = len(matrix[0])

        # First pass
        for i in range(r):
            for j in range(c):

                if matrix[i][j] == 0:
                    self.markInfinity(matrix, i, j)

        # Second pass
        for i in range(r):
            for j in range(c):

                if matrix[i][j] == float("inf"):
                    matrix[i][j] = 0
````


### ⏱ Complexity

```text
Time Complexity:
O((m × n) × (m + n))

Space Complexity:
O(1)
```
### 🧠 METHOD 2: Better Solution (Row & Column Tracking)

---

### 📌 Approach

Use:

```text
rowTrack[]
colTrack[]
```

Steps:

1. Track rows having zero
2. Track columns having zero
3. Set values to zero

---



### 💻 Code


````python
from typing import List

class Solution:

    def setZeroes(self, matrix: List[List[int]]) -> None:

        r = len(matrix)
        c = len(matrix[0])

        rowTrack = [0] * r
        colTrack = [0] * c

        # First pass
        for i in range(r):
            for j in range(c):

                if matrix[i][j] == 0:
                    rowTrack[i] = -1
                    colTrack[j] = -1

        # Second pass
        for i in range(r):
            for j in range(c):

                if rowTrack[i] == -1 or colTrack[j] == -1:
                    matrix[i][j] = 0
````


### ⏱ Complexity

```text
Time Complexity:
O(m × n)

Space Complexity:
O(m + n)
```

---



### 🧠 METHOD 3: Optimal (Using Matrix as Marker)



🔥 **Most Important Interview Method**

---

### 📌 Approach

Key Trick:

```text
Use first row & first column as markers
```

Steps:

1. Use first row/column as markers
2. Use `col0` variable for first column
3. Apply markers
4. Update first row & column

---



### 💻 Code


````python
from typing import List

class Solution:

    def setZeroes(self, matrix: List[List[int]]) -> None:

        r = len(matrix)
        c = len(matrix[0])

        col0 = 1

        # First pass
        for i in range(r):
            for j in range(c):

                if matrix[i][j] == 0:

                    if j == 0:
                        col0 = 0
                    else:
                        matrix[0][j] = 0
                        matrix[i][0] = 0

        # Second pass
        for i in range(1, r):
            for j in range(1, c):

                if matrix[0][j] == 0 or matrix[i][0] == 0:
                    matrix[i][j] = 0

        # First row
        if matrix[0][0] == 0:

            for j in range(c):
                matrix[0][j] = 0

        # First column
        if col0 == 0:

            for i in range(r):
                matrix[i][0] = 0
````


### ⏱ Complexity

```text
Time Complexity:
O(m × n)

Space Complexity:
O(1)   ⭐ Optimal
```

### ✅ LeetCode Style Implementation (Final Clean Code)


````python
"""
Problem: Set Matrix Zeroes
LeetCode: 73
Approach: Use first row & column as markers
Time Complexity: O(m × n)
Space Complexity: O(1)
"""

from typing import List

class Solution:

    def setZeroes(self, matrix: List[List[int]]) -> None:

        r = len(matrix)
        c = len(matrix[0])

        col0 = 1

        # Step 1: Mark
        for i in range(r):
            for j in range(c):

                if matrix[i][j] == 0:

                    if j == 0:
                        col0 = 0
                    else:
                        matrix[0][j] = 0
                        matrix[i][0] = 0

        # Step 2: Apply markers
        for i in range(1, r):
            for j in range(1, c):

                if matrix[0][j] == 0 or matrix[i][0] == 0:
                    matrix[i][j] = 0

        # Step 3: First row
        if matrix[0][0] == 0:

            for j in range(c):
                matrix[0][j] = 0

        # Step 4: First column
        if col0 == 0:

            for i in range(r):
                matrix[i][0] = 0
````

# 📘 Rotate Image — LeetCode 48



🔗 **Problem Link:**
[https://leetcode.com/problems/rotate-image/](https://leetcode.com/problems/rotate-image/)

📚 **Source Reference:**
[https://codeanddebug.in/blog/rotate-image-leetcode-48/](https://codeanddebug.in/blog/rotate-image-leetcode-48/)

---

### 🧩 Problem Statement

You are given an **n × n matrix** representing an image.

Rotate the image **90 degrees clockwise**.

⚠️ You must do this **in-place** (without using another matrix).

---

### 🧪 Example

```text
Input:
matrix =
[
 [1,2,3],
 [4,5,6],
 [7,8,9]
]

Output:
[
 [7,4,1],
 [8,5,2],
 [9,6,3]
]
```

---

### 🔒 Constraints

```text
n == matrix.length
n == matrix[i].length

1 <= n <= 20
-1000 <= matrix[i][j] <= 1000
```

---

### 🧠 METHOD 1: Brute Force (Using Extra Matrix)

---

### 📌 Approach

Key Idea:

```text
New Position Rule:

(row, col)
→
(col, n-1-row)
```

Steps:

1. Create empty matrix
2. Move each element to new position
3. Copy result back to original matrix

This method is **easy to understand**, but uses **extra space**.

---



### 💻 Code


````python
from typing import List

class Solution:

    def rotate(self, matrix: List[List[int]]) -> None:

        r = len(matrix)
        c = len(matrix[0])

        # Create new matrix
        result = [[0 for _ in range(c)] for _ in range(r)]

        # Fill rotated positions
        for i in range(r):
            for j in range(c):

                result[j][r - 1 - i] = matrix[i][j]

        # Copy back
        matrix[:] = result
````


---

### ⏱ Complexity

```text
Time Complexity:
O(n²)

Space Complexity:
O(n²)
```

---



### 🧠 METHOD 2: Optimal (Transpose + Reverse) ⭐



🔥 **Most Important Interview Method**

---

### 📌 Approach

Rotate matrix in **2 steps**:

---

### Step 1: Transpose Matrix

Swap:

```text
matrix[i][j] ↔ matrix[j][i]
```

Convert:

```text
Row → Column
```

Example:

```text
Before:

1 2 3
4 5 6
7 8 9

After Transpose:

1 4 7
2 5 8
3 6 9
```

---

### Step 2: Reverse Each Row

Convert:

```text
1 4 7
→
7 4 1
```

Final:

```text
7 4 1
8 5 2
9 6 3
```



### 💻 Code


````python
from typing import List

class Solution:

    def rotate(self, matrix: List[List[int]]) -> None:

        n = len(matrix)

        # Step 1: Transpose
        for i in range(n - 1):
            for j in range(i + 1, n):

                matrix[i][j], matrix[j][i] = \
                matrix[j][i], matrix[i][j]

        # Step 2: Reverse rows
        for i in range(n):

            matrix[i].reverse()
````


### ⏱ Complexity

```text
Time Complexity:
O(n²)

Space Complexity:
O(1)   ⭐ Optimal
```



### 🧠 METHOD 3: Layer-by-Layer Rotation (Cycle Rotation)



(Advanced alternative method)

---

### 📌 Approach

Rotate **one square layer at a time**.

Swap **4 elements at once**.

Pattern:

```text
top → right
right → bottom
bottom → left
left → top
```

Repeat layer by layer.

---



### 💻 Code


````python
from typing import List

class Solution:

    def rotate(self, matrix: List[List[int]]) -> None:

        n = len(matrix)

        left = 0
        right = n - 1

        while left < right:

            for i in range(right - left):

                top = left
                bottom = right

                # Save top-left
                temp = matrix[top][left + i]

                # bottom-left → top-left
                matrix[top][left + i] = \
                matrix[bottom - i][left]

                # bottom-right → bottom-left
                matrix[bottom - i][left] = \
                matrix[bottom][right - i]

                # top-right → bottom-right
                matrix[bottom][right - i] = \
                matrix[top + i][right]

                # temp → top-right
                matrix[top + i][right] = temp

            left += 1
            right -= 1
````


### ⏱ Complexity



```text
Time Complexity:
O(n²)

Space Complexity:
O(1)
```

# 📘 Spiral Matrix — LeetCode 54



🔗 **Problem Link:**
[https://leetcode.com/problems/spiral-matrix/](https://leetcode.com/problems/spiral-matrix/)

📚 **Source Reference:**
[https://codeanddebug.in/blog/spiral-matrix-leetcode-54/](https://codeanddebug.in/blog/spiral-matrix-leetcode-54/)

---

### 🧩 Problem Statement

Given an **m × n matrix**, return all elements of the matrix in **spiral order** (clockwise), starting from the **top-left corner**.

---

### 🧪 Example

```text
Input:
matrix =
[
 [1, 2, 3],
 [4, 5, 6],
 [7, 8, 9]
]

Output:
[1, 2, 3, 6, 9, 8, 7, 4, 5]
```

---

### 🔒 Constraints

```text
m == matrix.length
n == matrix[i].length
1 <= m, n <= 10
-100 <= matrix[i][j] <= 100
```

---



### 🧠 METHOD 1: Pointer Shrinking Technique (Optimal)


### 📌 Approach

This method uses **four boundary pointers**:

```text
top → first row  
bottom → last row  
left → first column  
right → last column  
```

Steps:

1. Traverse **Left ➜ Right** across top row
2. Traverse **Top ➜ Bottom** down right column
3. Traverse **Right ➜ Left** across bottom row
4. Traverse **Bottom ➜ Top** up left column
5. Shrink boundaries
6. Repeat until boundaries cross

This is called **Layer-by-Layer Traversal** (onion peeling).



### 💻 Code


````python
from typing import List

class Solution:
    def spiralOrder(self, matrix: List[List[int]]) -> List[int]:

        if not matrix or not matrix[0]:
            return []

        result = []

        top = 0
        bottom = len(matrix) - 1
        left = 0
        right = len(matrix[0]) - 1

        while top <= bottom and left <= right:

            # Left ➜ Right
            for i in range(left, right + 1):
                result.append(matrix[top][i])
            top += 1

            # Top ➜ Bottom
            for i in range(top, bottom + 1):
                result.append(matrix[i][right])
            right -= 1

            # Right ➜ Left
            if top <= bottom:
                for i in range(right, left - 1, -1):
                    result.append(matrix[bottom][i])
                bottom -= 1

            # Bottom ➜ Top
            if left <= right:
                for i in range(bottom, top - 1, -1):
                    result.append(matrix[i][left])
                left += 1

        return result
````


### ⏱ Time and Space Complexity

```text
Time Complexity:  O(m × n)

Every element is visited exactly once.

Space Complexity: O(1)

Only boundary pointers are used.
(Result list is not counted.)
```

---



### 🧠 METHOD 2: Simulation Using Visited Matrix

(Alternate but less optimal)

---

### 📌 Approach

1. Create **visited matrix**
2. Move in **4 directions**:

   * Right
   * Down
   * Left
   * Up
3. Change direction when:

   * Boundary reached
   * Cell already visited

---



### 💻 Code



````python
from typing import List

class Solution:
    def spiralOrder(self, matrix: List[List[int]]) -> List[int]:

        if not matrix:
            return []

        m = len(matrix)
        n = len(matrix[0])

        visited = [[False] * n for _ in range(m)]

        directions = [(0,1),(1,0),(0,-1),(-1,0)]

        dir_index = 0

        row = 0
        col = 0

        result = []

        for _ in range(m*n):

            result.append(matrix[row][col])

            visited[row][col] = True

            next_row = row + directions[dir_index][0]
            next_col = col + directions[dir_index][1]

            if (
                next_row < 0 or
                next_row >= m or
                next_col < 0 or
                next_col >= n or
                visited[next_row][next_col]
            ):
                dir_index = (dir_index + 1) % 4

            row += directions[dir_index][0]
            col += directions[dir_index][1]

        return result
````


---

### ⏱ Time and Space Complexity

```text
Time Complexity: O(m × n)

Space Complexity: O(m × n)

Extra visited matrix used.
```

---



### 🧠 METHOD 3: Recursive Layer Traversal



### 📌 Approach

1. Process outer layer
2. Recursively process inner matrix
3. Stop when boundaries cross

---



### 💻 Code

````python
from typing import List

class Solution:

    def spiralOrder(self, matrix: List[List[int]]) -> List[int]:

        result = []

        def spiral(top, bottom, left, right):

            if top > bottom or left > right:
                return

            # top row
            for i in range(left, right + 1):
                result.append(matrix[top][i])

            # right column
            for i in range(top + 1, bottom + 1):
                result.append(matrix[i][right])

            if top < bottom:

                # bottom row
                for i in range(right - 1, left - 1, -1):
                    result.append(matrix[bottom][i])

            if left < right:

                # left column
                for i in range(bottom - 1, top, -1):
                    result.append(matrix[i][left])

            spiral(
                top + 1,
                bottom - 1,
                left + 1,
                right - 1
            )

        spiral(
            0,
            len(matrix) - 1,
            0,
            len(matrix[0]) - 1
        )

        return result
````


---

### ⏱ Time and Space Complexity

```text
Time Complexity: O(m × n)

Space Complexity: O(min(m,n))

Due to recursive call stack.
```
