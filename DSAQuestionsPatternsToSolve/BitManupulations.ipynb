# Question 1 → Check Odd Or Even

```python id="9x2kpm"
def odd_even(n):

    if n & 1:

        return "Odd"

    return "Even"


print(odd_even(7))
```

### Insight

1. **Last Bit Decide Odd/Even**
2. **Odd Number Last Bit = 1**
3. **Even Number Last Bit = 0**
4. **Bitwise AND Check Last Bit Quickly**

---

# Important Odd/Even Formula

n & 1

---

# Question 2 → Find Unique Element Using XOR

```python id="6m8v0q"
def unique_element(arr):

    ans = 0

    for num in arr:

        ans ^= num

    return ans


arr = [2,3,5,4,5,3,4]

print(unique_element(arr))
```

### Insight

1. **Same Numbers Cancel Each Other**
2. **XOR Of Same Numbers = 0**
3. **0 XOR Number = Number**
4. **Only Unique Element Remains**
5. **Continuous XOR Updating Happens**

# Question 3 → Check Power Of Two

```python id="1v4zkr"
def power_of_two(n):

    if n <= 0:
        return False

    return (n & (n - 1)) == 0


print(power_of_two(16))
```

### Insight

1. **Power Of Two Has Only One Set Bit**
2. **n-1 Flips Bits After That Set Bit**
3. **AND Operation Removes Common Bits**
4. **Result Becomes Zero**
5. **Efficient Binary Trick**

---

# Important Power Of Two Formula

n & (n-1) = 0

---

# Question 4 → Count Set Bits

```python id="5q1xmn"
def count_set_bits(n):

    count = 0

    while n:

        count += n & 1

        n >>= 1

    return count


print(count_set_bits(13))
```

### Insight

1. **Check Last Bit**
2. **If Last Bit = 1 → Increase Count**
3. **Right Shift Removes Last Bit**
4. **Repeat Until Number Becomes Zero**
5. **Binary Traversal Happens Bit By Bit**

# Question 5 → Swap Two Numbers Using XOR

```python id="7k0vpa"
def swap(a, b):

    a = a ^ b

    b = a ^ b

    a = a ^ b

    return a, b


print(swap(5, 10))
```

### Insight

1. **XOR Helps Store Difference**
2. **No Extra Variable Needed**
3. **Values Exchange Using XOR Properties**
4. **Bit Manipulation Performs Swapping Efficiently**
