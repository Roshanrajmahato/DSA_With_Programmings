# Question 1 → K Largest Elements

```python id="7x2nva"
import heapq

def k_largest(arr, k):

    min_heap = []

    for num in arr:

        heapq.heappush(min_heap, num)

        if len(min_heap) > k:

            heapq.heappop(min_heap)

    return min_heap


arr = [3,10,5,20,15,7]

print(k_largest(arr, 3))
```

### Insight

1. **Maintain Small Heap Of Size K**
2. **Heap Stores Best K Elements**
3. **If Size Exceeds K → Remove Smallest**
4. **Final Heap Contains K Largest Elements**

---

# Question 2 → Kth Smallest Element

```python id="w8q1rb"
import heapq

def kth_smallest(arr, k):

    heapq.heapify(arr)

    for _ in range(k - 1):

        heapq.heappop(arr)

    return heapq.heappop(arr)


arr = [7,10,4,3,20,15]

print(kth_smallest(arr, 3))
```

### Insight

1. **Heap Automatically Keeps Smallest At Top**
2. **Remove Smaller Elements One By One**
3. **Kth Removed Element Become Answer**
4. **Priority Processing Happens Automatically**

# Question 3 → Merge K Sorted Lists

```python id="0g3yzn"
import heapq

def merge_k_lists(lists):

    heap = []

    for i in range(len(lists)):

        if lists[i]:

            heapq.heappush(heap, (lists[i][0], i, 0))

    result = []

    while heap:

        value, row, col = heapq.heappop(heap)

        result.append(value)

        if col + 1 < len(lists[row]):

            heapq.heappush(
                heap,
                (lists[row][col + 1], row, col + 1)
            )

    return result


lists = [
    [1,4,5],
    [1,3,4],
    [2,6]
]

print(merge_k_lists(lists))
```

### Insight

1. **Heap Stores Current Smallest Element**
2. **Take Minimum From All Lists**
3. **After Removing One Element**
4. **Insert Next Element From Same List**
5. **Heap Maintains Global Minimum**

---

# Question 4 → Top K Frequent Elements

```python id="7n4jyp"
import heapq
from collections import Counter

def top_k_frequent(arr, k):

    freq = Counter(arr)

    heap = []

    for num, count in freq.items():

        heapq.heappush(heap, (-count, num))

    result = []

    for _ in range(k):

        result.append(heapq.heappop(heap)[1])

    return result


arr = [1,1,1,2,2,3]

print(top_k_frequent(arr, 2))
```

### Insight

1. **Count Frequency First**
2. **Higher Frequency Means Higher Priority**
3. **Use Max Heap Logic**
4. **Most Frequent Elements Come First**
5. **Priority Queue Maintains Best Candidates**

# Question 5 → Connect Ropes With Minimum Cost

```python id="z1p8mk"
import heapq

def connect_ropes(arr):

    heapq.heapify(arr)

    total_cost = 0

    while len(arr) > 1:

        first = heapq.heappop(arr)

        second = heapq.heappop(arr)

        cost = first + second

        total_cost += cost

        heapq.heappush(arr, cost)

    return total_cost


arr = [4,3,2,6]

print(connect_ropes(arr))
```

### Insight

1. **Always Connect Smallest Ropes First**
2. **Small Cost Early Helps Reduce Future Cost**
3. **Heap Gives Minimum Ropes Quickly**
4. **New Rope Insert Back Into Heap**
5. **Greedy + Heap Combination Used**
