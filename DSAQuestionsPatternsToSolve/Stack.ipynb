# Question 1 → Valid Parentheses

```python id="rqm4zq"
def valid_parentheses(s):

    stack = []

    mapping = {
        ')': '(',
        '}': '{',
        ']': '['
    }

    for ch in s:

        if ch in "({[":
            stack.append(ch)

        else:

            if not stack:
                return False

            top = stack.pop()

            if mapping[ch] != top:
                return False

    return len(stack) == 0


s = "{[()]}"

print(valid_parentheses(s))
```

### Insight

1. **Opening Bracket Store In Stack**
2. **Closing Bracket Check Recent Opening**
3. **Top Element Must Match**
4. **If Not Match → Invalid**
5. **At End Stack Must Become Empty**

---

# Question 2 → Reverse String Using Stack

```python id="m5tt05"
def reverse_string(s):

    stack = []

    for ch in s:
        stack.append(ch)

    ans = ""

    while stack:
        ans += stack.pop()

    return ans


s = "ROSHAN"

print(reverse_string(s))
```

### Insight

1. **Push All Characters**
2. **Pop One By One**
3. **Because Stack Works Reverse**
4. **String Automatically Reverse**

# Question 3 → Next Greater Element

```python id="n2cg0k"
def next_greater(arr):

    stack = []

    result = [-1] * len(arr)

    for i in range(len(arr)-1, -1, -1):

        while stack and stack[-1] <= arr[i]:
            stack.pop()

        if stack:
            result[i] = stack[-1]

        stack.append(arr[i])

    return result


arr = [4,5,2,10,8]

print(next_greater(arr))
```

### Insight

1. **Need Next Bigger Element**
2. **Remove Smaller Elements From Stack**
3. **Top Element Become Answer**
4. **Current Element Store For Future**
5. **Stack Maintain Useful Elements Only**

---

# Question 4 → Min Stack

```python id="1f7rkk"
class MinStack:

    def __init__(self):

        self.stack = []
        self.min_stack = []

    def push(self, val):

        self.stack.append(val)

        if not self.min_stack or val <= self.min_stack[-1]:
            self.min_stack.append(val)

    def pop(self):

        if self.stack[-1] == self.min_stack[-1]:
            self.min_stack.pop()

        self.stack.pop()

    def get_min(self):

        return self.min_stack[-1]


obj = MinStack()

obj.push(5)
obj.push(2)
obj.push(8)
obj.push(1)

print(obj.get_min())
```

### Insight

1. **One Stack Store Normal Elements**
2. **Another Stack Store Minimum Elements**
3. **Whenever Smaller Element Comes → Store**
4. **Top Of Min Stack Always Current Minimum**

# Question 5 → Stock Span Problem

```python id="0hzyt5"
def stock_span(prices):

    stack = []

    span = []

    for i in range(len(prices)):

        while stack and prices[stack[-1]] <= prices[i]:
            stack.pop()

        if not stack:
            span.append(i + 1)

        else:
            span.append(i - stack[-1])

        stack.append(i)

    return span


prices = [100,80,60,70,60,75,85]

print(stock_span(prices))
```

### Insight

1. **Need Previous Greater Price**
2. **Remove Smaller Previous Prices**
3. **Nearest Greater Element Help Calculate Span**
4. **Stack Store Useful Indexes Only**
5. **Span Continous Updating**
