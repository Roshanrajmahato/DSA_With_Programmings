# Question 1 → Inorder Traversal

```python id="j7m0vl"
class Node:

    def __init__(self, val):

        self.val = val
        self.left = None
        self.right = None


def inorder(root):

    if root is None:
        return

    inorder(root.left)

    print(root.val, end=" ")

    inorder(root.right)


root = Node(1)

root.left = Node(2)
root.right = Node(3)

root.left.left = Node(4)
root.left.right = Node(5)

inorder(root)
```

### Insight

1. **Traverse Left First**
2. **Then Visit Current Node**
3. **Then Traverse Right**
4. **Recursive Calls Handle Subtrees**
5. **Left → Root → Right Pattern**

---

# Question 2 → Height Of Binary Tree

```python id="vq1r5a"
def height(root):

    if root is None:
        return 0

    left_height = height(root.left)

    right_height = height(root.right)

    return 1 + max(left_height, right_height)
```

### Insight

1. **Height Depends On Child Heights**
2. **Find Left Height**
3. **Find Right Height**
4. **Take Maximum**
5. **Add Current Node**
6. **Recursive Bottom-Up Calculation**

---

# Question 3 → Count Total Nodes

```python id="h2v8nx"
def count_nodes(root):

    if root is None:
        return 0

    left_count = count_nodes(root.left)

    right_count = count_nodes(root.right)

    return 1 + left_count + right_count
```

### Insight

1. **Current Total Depends On Subtrees**
2. **Count Left Nodes**
3. **Count Right Nodes**
4. **Add Current Node**
5. **Recursive Division Happens Automatically**

---

# Question 4 → Check Identical Trees

```python id="8j3w9k"
def identical(root1, root2):

    if root1 is None and root2 is None:
        return True

    if root1 is None or root2 is None:
        return False

    return (
        root1.val == root2.val
        and identical(root1.left, root2.left)
        and identical(root1.right, root2.right)
    )
```

### Insight

1. **Compare Current Nodes**
2. **Then Compare Left Subtrees**
3. **Then Compare Right Subtrees**
4. **All Conditions Must Match**
5. **Structure + Values Both Important**

# Question 5 → Lowest Common Ancestor (LCA)

```python id="kk9x2m"
def lca(root, p, q):

    if root is None:
        return None

    if root.val == p or root.val == q:
        return root

    left = lca(root.left, p, q)

    right = lca(root.right, p, q)

    if left and right:
        return root

    return left if left else right
```

### Insight

1. **Search Both Nodes In Subtrees**
2. **If One Found Left & One Found Right**
3. **Current Node Become LCA**
4. **Recursive Search Happens Bottom-Up**
5. **First Common Meeting Point Is Answer**
