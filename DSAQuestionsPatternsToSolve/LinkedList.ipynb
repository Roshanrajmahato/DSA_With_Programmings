# Question 1 → Traversal Of Linked List

```python id="8m3ywa"
class Node:

    def __init__(self, data):

        self.data = data
        self.next = None


def traversal(head):

    temp = head

    while temp:

        print(temp.data, end=" -> ")

        temp = temp.next


head = Node(10)

head.next = Node(20)
head.next.next = Node(30)

traversal(head)
```

### Insight

1. **Start From Head**
2. **Visit Current Node**
3. **Move Using Next Pointer**
4. **Continue Until NULL**
5. **Sequential Traversal Happens**

---

# Question 2 → Reverse Linked List

```python id="8w2k9q"
class Node:

    def __init__(self, data):

        self.data = data
        self.next = None


def reverse(head):

    prev = None

    current = head

    while current:

        next_node = current.next

        current.next = prev

        prev = current

        current = next_node

    return prev
```

### Insight

1. **Need To Reverse Direction Of Links**
2. **Store Next Node First**
3. **Reverse Current Link**
4. **Move Previous & Current Forward**
5. **Pointer Manipulation Main Logic**

# Question 3 → Detect Cycle In Linked List

```python id="3j0qmv"
def detect_cycle(head):

    slow = head

    fast = head

    while fast and fast.next:

        slow = slow.next

        fast = fast.next.next

        if slow == fast:
            return True

    return False
```

### Insight

1. **Slow Pointer Moves One Step**
2. **Fast Pointer Moves Two Steps**
3. **If Cycle Exists → Both Meet**
4. **If Fast Reaches NULL → No Cycle**
5. **Floyd Cycle Detection Logic**

---

# Important Slow/Fast Formula

Slow = 1\ Step \quad Fast = 2\ Steps

---

# Question 4 → Find Middle Of Linked List

```python id="2d9x0r"
def middle_node(head):

    slow = head

    fast = head

    while fast and fast.next:

        slow = slow.next

        fast = fast.next.next

    return slow.data
```

### Insight

1. **Slow Moves One Step**
2. **Fast Moves Two Steps**
3. **When Fast Reaches End**
4. **Slow Automatically Reaches Middle**
5. **Two Pointer Optimization**

# Question 5 → Merge Two Sorted Linked Lists

```python id="9q1wxt"
class Node:

    def __init__(self, data):

        self.data = data
        self.next = None


def merge(l1, l2):

    dummy = Node(-1)

    tail = dummy

    while l1 and l2:

        if l1.data < l2.data:

            tail.next = l1

            l1 = l1.next

        else:

            tail.next = l2

            l2 = l2.next

        tail = tail.next

    if l1:
        tail.next = l1

    if l2:
        tail.next = l2

    return dummy.next
```

### Insight

1. **Compare Both Current Nodes**
2. **Smaller Node Attach First**
3. **Move That List Forward**
4. **Tail Pointer Builds Final List**
5. **Sorted Order Maintained Automatically**
