# Question 1 → DFS Traversal

```python id="o4q7zm"
def dfs(graph, node, visited):

    visited.add(node)

    print(node, end=" ")

    for neighbor in graph[node]:

        if neighbor not in visited:

            dfs(graph, neighbor, visited)


graph = {
    0:[1,2],
    1:[0,3],
    2:[0,4],
    3:[],
    4:[]
}

visited = set()

dfs(graph, 0, visited)
```

### Insight

1. **Visit Current Node**
2. **Mark As Visited**
3. **Explore One Path Deeply**
4. **Recursive Calls Traverse Graph**
5. **Backtrack After Path Ends**

---

# Question 2 → BFS Traversal

```python id="7k2v0r"
from collections import deque

def bfs(graph, start):

    visited = set([start])

    q = deque([start])

    while q:

        node = q.popleft()

        print(node, end=" ")

        for neighbor in graph[node]:

            if neighbor not in visited:

                visited.add(neighbor)

                q.append(neighbor)


graph = {
    0:[1,2],
    1:[3],
    2:[4],
    3:[],
    4:[]
}

bfs(graph, 0)
```

### Insight

1. **Visit Nearest Nodes First**
2. **Queue Maintain Processing Order**
3. **Current Node Explore All Neighbors**
4. **Level By Level Traversal Happens**
5. **Visited Prevents Repeating**

# Question 3 → Detect Cycle In Undirected Graph

```python id="0w9mxa"
def detect_cycle(graph, node, visited, parent):

    visited.add(node)

    for neighbor in graph[node]:

        if neighbor not in visited:

            if detect_cycle(graph, neighbor, visited, node):
                return True

        elif neighbor != parent:
            return True

    return False


graph = {
    0:[1],
    1:[0,2],
    2:[1,3],
    3:[2,1]
}

visited = set()

print(detect_cycle(graph, 0, visited, -1))
```

### Insight

1. **Traverse Graph Using DFS**
2. **Track Parent Node**
3. **If Already Visited Node Found Again**
4. **And Not Parent → Cycle Exists**
5. **Visited Tracking Very Important**

---

# Question 4 → Number Of Connected Components

```python id="6t1r8p"
def dfs(graph, node, visited):

    visited.add(node)

    for neighbor in graph[node]:

        if neighbor not in visited:

            dfs(graph, neighbor, visited)


def connected_components(graph):

    visited = set()

    count = 0

    for node in graph:

        if node not in visited:

            dfs(graph, node, visited)

            count += 1

    return count


graph = {
    0:[1],
    1:[0],
    2:[3],
    3:[2],
    4:[]
}

print(connected_components(graph))
```

### Insight

1. **Each DFS Traversal Covers One Component**
2. **Unvisited Node Means New Component**
3. **Start DFS From That Node**
4. **Increase Component Count**
5. **Visited Prevents Duplicate Traversal**

# Question 5 → Shortest Path Using BFS

```python id="6l2h9z"
from collections import deque

def shortest_path(graph, start):

    distance = {start: 0}

    q = deque([start])

    while q:

        node = q.popleft()

        for neighbor in graph[node]:

            if neighbor not in distance:

                distance[neighbor] = distance[node] + 1

                q.append(neighbor)

    return distance


graph = {
    0:[1,2],
    1:[3],
    2:[3],
    3:[4],
    4:[]
}

print(shortest_path(graph, 0))
```

### Insight

1. **BFS Always Reaches Nearest Path First**
2. **Store Distance Of Each Node**
3. **Current Node Distance Helps Neighbor Distance**
4. **Queue Process Nodes Level By Level**
5. **First Time Visiting Gives Shortest Path**
