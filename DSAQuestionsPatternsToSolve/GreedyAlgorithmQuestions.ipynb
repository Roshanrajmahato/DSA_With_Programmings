# Question 1 → Activity Selection Problem

```python id="v4n0rk"
def activity_selection(start, end):

    activities = list(zip(start, end))

    activities.sort(key=lambda x: x[1])

    count = 1

    last_end = activities[0][1]

    for i in range(1, len(activities)):

        if activities[i][0] >= last_end:

            count += 1

            last_end = activities[i][1]

    return count


start = [1,3,0,5,8,5]
end = [2,4,6,7,9,9]

print(activity_selection(start, end))
```

### Insight

1. **Choose Activity That Finishes Earliest**
2. **More Remaining Time For Future Activities**
3. **Skip Overlapping Activities**
4. **Current Best Choice Helps Maximum Selection**

---

# Question 2 → Minimum Coins Problem

```python id="8b2nmt"
def minimum_coins(coins, amount):

    coins.sort(reverse=True)

    count = 0

    for coin in coins:

        while amount >= coin:

            amount -= coin

            count += 1

    return count


coins = [1,2,5,10]

print(minimum_coins(coins, 28))
```

### Insight

1. **Always Take Biggest Coin First**
2. **Reduce Remaining Amount Quickly**
3. **Repeat Until Amount Becomes Zero**
4. **Greedy Choice Reduce Number Of Coins**

# Question 3 → Fractional Knapsack

```python id="5mf3jv"
def fractional_knapsack(weights, values, capacity):

    items = []

    for i in range(len(weights)):

        ratio = values[i] / weights[i]

        items.append((ratio, weights[i], values[i]))

    items.sort(reverse=True)

    total_value = 0

    for ratio, weight, value in items:

        if capacity >= weight:

            total_value += value

            capacity -= weight

        else:

            total_value += ratio * capacity

            break

    return total_value


weights = [10,20,30]
values = [60,100,120]

print(fractional_knapsack(weights, values, 50))
```

### Insight

1. **Profit Per Weight Is Important**
2. **Take Highest Ratio First**
3. **Fill Bag Greedily**
4. **If Full Item Not Possible → Take Fraction**
5. **Current Best Ratio Gives Maximum Profit**

---

# Important Knapsack Formula

Ratio = \frac{Value}{Weight}

---

# Question 4 → Jump Game

```python id="5t12n9"
def can_jump(nums):

    max_reach = 0

    for i in range(len(nums)):

        if i > max_reach:
            return False

        max_reach = max(max_reach, i + nums[i])

    return True


nums = [2,3,1,1,4]

print(can_jump(nums))
```

### Insight

1. **Track Farthest Reachable Position**
2. **Current Position Must Be Reachable**
3. **Update Maximum Reach**
4. **Greedy Decision Helps Move Forward**
5. **If Position Unreachable → Answer False**

# Question 5 → Job Sequencing Problem

```python id="9n1k2x"
def job_sequencing(jobs):

    jobs.sort(key=lambda x: x[2], reverse=True)

    max_deadline = max(job[1] for job in jobs)

    slots = [-1] * (max_deadline + 1)

    profit = 0

    for job in jobs:

        job_id, deadline, gain = job

        for j in range(deadline, 0, -1):

            if slots[j] == -1:

                slots[j] = job_id

                profit += gain

                break

    return profit


jobs = [
    ('A',2,100),
    ('B',1,19),
    ('C',2,27),
    ('D',1,25),
    ('E',3,15)
]

print(job_sequencing(jobs))
```

### Insight

1. **Higher Profit Job Is More Important**
2. **Sort Jobs By Profit**
3. **Try Latest Possible Empty Slot**
4. **Avoid Wasting Early Slots**
5. **Greedy Choice Maximizes Total Profit**
