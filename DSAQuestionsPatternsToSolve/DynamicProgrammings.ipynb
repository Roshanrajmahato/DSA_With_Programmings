# Question 1 → Fibonacci Using DP

```python id="3o5k9x"
def fibonacci(n):

    dp = [0] * (n + 1)

    dp[1] = 1

    for i in range(2, n + 1):

        dp[i] = dp[i-1] + dp[i-2]

    return dp[n]


print(fibonacci(6))
```

### Insight

1. **Each Fibonacci Depends On Previous Two**
2. **Store Previous Answers**
3. **Avoid Recursive Recalculation**
4. **DP Array Continous Updating**

---

# Important Fibonacci Formula

dp[i]=dp[i-1]+dp[i-2]

---

# Question 2 → Climbing Stairs

```python id="7i1pzc"
def climb_stairs(n):

    dp = [0] * (n + 1)

    dp[0] = 1
    dp[1] = 1

    for i in range(2, n + 1):

        dp[i] = dp[i-1] + dp[i-2]

    return dp[n]


print(climb_stairs(5))
```

### Insight

1. **To Reach Current Step**
2. **Can Come From One Step Before**
3. **OR Two Steps Before**
4. **Current Ways Depend On Previous Ways**
5. **Store Answers In DP Array**

# Question 3 → Maximum Sum Non Adjacent Elements

```python id="aq0v7w"
def max_non_adjacent(arr):

    n = len(arr)

    dp = [0] * n

    dp[0] = arr[0]

    dp[1] = max(arr[0], arr[1])

    for i in range(2, n):

        take = arr[i] + dp[i-2]

        not_take = dp[i-1]

        dp[i] = max(take, not_take)

    return dp[-1]


arr = [2,7,9,3,1]

print(max_non_adjacent(arr))
```

### Insight

1. **At Every Element Two Choices**
2. **Take Current Element**
3. **OR Skip Current Element**
4. **If Take → Cannot Take Adjacent**
5. **Choose Maximum Possible Result**
6. **Previous Answers Help Current Decision**

---

# Question 4 → Coin Change (Minimum Coins)

```python id="t3v7mn"
def coin_change(coins, amount):

    dp = [float('inf')] * (amount + 1)

    dp[0] = 0

    for i in range(1, amount + 1):

        for coin in coins:

            if i - coin >= 0:

                dp[i] = min(dp[i], 1 + dp[i - coin])

    return dp[amount] if dp[amount] != float('inf') else -1


coins = [1,2,5]

print(coin_change(coins, 11))
```

### Insight

1. **Current Amount Depends On Smaller Amount**
2. **Try Every Coin**
3. **Take Minimum Coins Possible**
4. **Previous Smaller Solutions Reused**
5. **DP Array Store Best Answer**

# Question 5 → Longest Common Subsequence (LCS)

```python id="j0vlc5"
def lcs(text1, text2):

    n = len(text1)
    m = len(text2)

    dp = [[0]*(m+1) for _ in range(n+1)]

    for i in range(1, n+1):

        for j in range(1, m+1):

            if text1[i-1] == text2[j-1]:

                dp[i][j] = 1 + dp[i-1][j-1]

            else:

                dp[i][j] = max(dp[i-1][j], dp[i][j-1])

    return dp[n][m]


text1 = "abcde"
text2 = "ace"

print(lcs(text1, text2))
```

### Insight

1. **Compare Characters One By One**
2. **If Match → Increase Length**
3. **Else Take Best Previous Answer**
4. **Current Cell Depends On Previous Cells**
5. **2D DP Table Continous Updating**
6. **Store Best Common Subsequence Length**
